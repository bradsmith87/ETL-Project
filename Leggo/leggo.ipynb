```python
import pandas as pd
from sqlalchemy import create_engine
from config import username

```

```python
colors_file = "Resources/colors.csv"
colors_df = pd.read_csv(colors_file)
colors_df.head()
```

```python
rename_color= colors_df.rename(columns={"id":"color_id","name":"color_name"})
rename_color.head()
```

```python
parts_file = "Resources/parts.csv"
parts_df = pd.read_csv(parts_file)
rename_parts_df= parts_df.rename(columns={"name":"part_name"})
rename_parts_df.head()
```

```python
inventory_parts_file = "Resources/inventory_parts.csv"
inventory_parts_df = pd.read_csv(inventory_parts_file)
inventory_parts_df.head()
```

```python
sets_file = "Resources/sets.csv"
sets_df=pd.read_csv(sets_file)
rename_sets_df= sets_df.rename(columns={"name":"set_name"})
rename_sets_df.head()
```

```python
inventories = "Resources/inventories.csv"
inventories_df=pd.read_csv(inventories)
inventories_df.head()
```

```python
rename_id= inventories_df.rename(columns={"id":"inventory_id"})
rename_id.head()
```

```python
sets_info= pd.merge(rename_id,rename_sets_df, on = 'set_num', how='outer').dropna()
sets_info.head()
```

```python
merge = pd.merge(inventory_parts_df,rename_color, on ="color_id", how="outer").dropna()
merge.head()
```

```python
inventory_info=pd.merge(merge,rename_parts_df, on="part_num",how="outer").dropna()
inventory_info.head()
```

```python
clean_data = pd.merge(sets_info,inventory_info, on="inventory_id",how="outer").dropna()
clean_data.head()
```

```python
export_csv = clean_data.to_csv("Resources/leggo.csv",index=False)
```
